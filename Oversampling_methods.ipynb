## RandomOverSampler

```python
from collections import Counter
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

X, y = make_classification(n_classes = 2, class_sep = 2,
                           weights = [0.1, 0.9], n_informative = 3,
                           n_redundant = 1, flip_y = 0,
                           n_features = 20, n_clusters_per_class = 1,
                           n_samples = 2000, random_state = 345)
counter = Counter(y)
```

```python
print(counter)
```

```python
# Scatter plot of given_dataset
for label, _ in counter.items():
    row_idx = np.where(y==label)[0]
    plt.scatter(X[row_idx, 0], X[row_idx, 1], label = str(label))
plt.legend()
plt.show()
```

```python
ros = RandomOverSampler(random_state = 42)
X_res, y_res = ros.fit_resample(X, y)
counter_ros = Counter(y_res)
print(counter_ros)

for label, _ in counter.items():
    row_idx = np.where(y==label)[0]
    plt.scatter(X_res[row_idx, 0], X_res[row_idx, 1], label = str(label))
plt.legend()
plt.show()
```

## SMOTE

```python
sm = SMOTE()
X_res2, y_res2 = sm.fit_resample(X, y)
counter_sm = Counter(y_res2)
print(counter_sm)

for label, _ in counter.items():
    row_idx = np.where(y==label)[0]
    plt.scatter(X_res2[row_idx, 0], X_res2[row_idx, 1], label = str(label))
plt.legend()
plt.show()
```
