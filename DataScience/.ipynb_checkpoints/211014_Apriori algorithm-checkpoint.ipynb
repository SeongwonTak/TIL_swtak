```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
store_df = pd.read_csv('C:/Users/USER/Desktop/TIL_swtak/DataScience/Dataset/store_data.csv', header = None)
```

```python
store_df
```

```python
records = []
for i in range(len(store_df)):
    records.append([str(store_df.values[i,j]) \
                    for j in range(len(store_df.columns)) \
                    if not pd.isna(store_df.values[i,j])])
```

```python
te = TransactionEncoder()
te_ary = te.fit(records).transform(records, sparse = True)
te_df = pd.DataFrame.sparse.from_spmatrix(te_ary, columns = te.columns_)
```

```python
te_df
```

```python
frequent_itemset = apriori(te_df,
                          min_support = 0.005,
                          max_len = 2,
                          use_colnames = True)
```

```python
frequent_itemset
```

```python
frequent_itemset['length'] = frequent_itemset['itemsets'].apply(lambda x : len(x))
```

```python
item_asso_list = frequent_itemset.loc[frequent_itemset.length >= 2].sort_values('support', ascending = False)
```

```python
item_asso_list
```

다음과 같이 길이 2 이상의 support 조합을 찾아낼 수 있다. 여기서 추가적으로 신뢰도가 특정 이상인지, 혹은 다른 측도를 추가로 컬럼에 붙일 수 있을 것이다.

```python
association_rules_df = association_rules(frequent_itemset, 
                                         metric='confidence', 
                                         min_threshold=0.005,
                                        )
```

위의 방식을 사용하면 다른 지표에 대해서도 표를 봅아볼 수 있다.


```python
association_rules_df.sort_values('confidence', ascending = False).iloc[:20,:]
```

우유와 물, 토마토소스와 스파게티 등 유의미한 조합이 발견되고 있다.

또한 lift를 통해 상관관계 또한 같이 확인가능해볼 것이다.

```python
association_rules_df.sort_values('lift', ascending = False).iloc[:20,:]
```

조금 더 상관관계 기준으로 높은 아이템 쌍들이 추천되고 있음을 알 수 있다.
다만 상대적으로 confidential 등이 낮은 경우도 보이는데, 종합적으로 고려하는 것이 유의미할것이다라고 추론해 볼 수 있을 것이다.
