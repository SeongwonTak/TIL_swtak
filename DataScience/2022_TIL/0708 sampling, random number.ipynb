# pandas - 랜덤 표본추출

```python
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
```

```python
boston_df
```

506개의 집에서 샘플링을 실시해보자.

```python
# 단순 무작위표본 추출, 비복원
boston_df.sample(n=5, replace = False)
```

```python
# 비율 설정 가능.
boston_df.sample(frac = 0.05)
```

```python
# weight 설정 가능
boston_df.sample(frac = 0.05, weights = 'CHAS')
```

weights는 값이 클수록, 더 높은 비중을 준다. 그렇기에, 0의 값을 가진 CHAS는 나오지 않았음.

```python
import random
# 계통 추출법 알고리즘
# n은 몇 step 건널것인가.
def sys_sampling(data, n):
    N = len(data)
    K = int(N/n) # sample 수
    index =  data[:K].sample(1).index
    result_df = pd.DataFrame()
    
    while len(result_df) <n:
        result_df = result_df.append(data.loc[index, :])
        index += K
    
    return result_df
```

```python
sys_sampling(boston_df, 5)
```

층화 추출의 경우는 scikit-learn의 stratify option을 활용하면 충분.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(boston_df.drop('CHAS', axis = 1), 
                                                    boston_df['CHAS'],
                                                    test_size = 0.1,
                                                    shuffle = True,
                                                    stratify = boston_df['CHAS'])
```

```python
pd.DataFrame(y_train).value_counts() / len(y_train)
```

```python
pd.DataFrame(y_test).value_counts() / len(y_test)
```
