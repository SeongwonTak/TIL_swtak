## 회귀 대체법

```python
df_reg = df.copy()
```

결측치가 없는 행을 바탕으로 모델을 작성 후, 추정을 진행한다.

```python
df_nonan = df.dropna(axis = 0)
X = df_nonan.drop('mean radius', axis = 1)
y = df_nonan[['mean radius']]

# 그리고 실제로 값을 입력할 전체 테스트본도 가져올 것이다.
X_true = df.drop('mean radius', axis = 1)
```

```python
# 회귀 진행
reg = LinearRegression()
model = reg.fit(X, y)
```

```python
y_pred = reg.predict(X_true)
```

이제 이를 대체하는 법을 알고자 한다. 물론 for문을 활용하여 돌리면 좋겠지만, 성능이 많이 떨어지므로 다른 방법을 써보자.

```python
# 결측값이 true인 경우에 y_pred_flatten의 값을 가져오고, false인 경우에는 기존 값을 가져와서 할당한다.
df_reg['mean radius'] = np.where(df_reg['mean radius'].isnull(),
                                 pd.Series(y_pred.flatten()),
                                 df_reg['mean radius'])
```

```python
df_reg
```

실제 참 값과 비교하면, 큰 차이는 존재하지 않는다.
