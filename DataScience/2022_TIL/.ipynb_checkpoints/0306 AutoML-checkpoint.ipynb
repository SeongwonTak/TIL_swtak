# AutoML의 구현 예제

데이터 성격에 맞게 자동으로 데이터 분석 모델을 추천해 주는 AutoML.

```python
#!pip install pycaret
```

```python
from pycaret.datasets import get_data
dataset = get_data('bank')
```

```python
dataset
```

```python
SEED = 42

# train, test에 대한 분리를 실시하자.
train_data = dataset.sample(frac = 0.8, random_state = SEED).reset_index(drop = True)
test_data = dataset.drop(train_data.index).reset_index(drop = True)
```

```python
train_data.shape
```

```python
test_data.shape
```

80 : 20 정도로 train / test를 나누었고, 이를 바탕으로 학습을 진행하고자 한다.

## 모델 기본 준비
deposit을 맞히는 문제로 고려하자.

```python
from pycaret.classification import *

model = setup(
    data = train_data,
    target = 'deposit'
)
```

학습을 위해 모델을 만드는 방법을 알아보자.

```python
compare_models(
    exclude = None,
    include = None,
    fold = 10,
    round = 4,
    sort = 'Accuracy',
    n_select = 3,
    turbo = True,
    verbose = True
)
```

정확도나, f1-score 면에서도 CatBoost가 가장 우수함을 알 수 있다. (시간적 이슈를 제외한다.)  
시간적 이슈까지 고려할 경우, LightGBM을 사용해도, 큰 정확도 손실이 없다.  

여기서는 catboost를 활용하는 상황을 보자.

```python
catboost = create_model('catboost')
```

```python
plot_model(catboost, plot = 'error')
```

```python
plot_model(catboost, plot = 'feature')
```

```python
plot_model(catboost)
```
